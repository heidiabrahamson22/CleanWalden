```python
#Question 1
counter = 0
with open("walden.txt", 'r', encoding='utf8') as file:
    for line in file:
        counter = counter + 1

print(counter)
```

```python
#Question 2
with open("walden.txt", "r") as file:
    walden_list = file.readlines()
walden_list[:15]
```

```python
#Question 3
for index, value in enumerate(walden_list):
    walden_list[index]=value.lower().replace(",","").replace(":","").replace(";","").strip()
walden_list[:30]
```

```python
#Question 4a
word_list = []
for line in walden_list:
    words = line.split(" ")
    word_list.append(words)
flat_list = []
for sublist in word_list:
    for item in sublist:
        flat_list.append(item)
count = {}
for word in flat_list:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1
len(count)
```

```python
#Question 4b
print("distasteful:"+ str(count.get("distasteful", 0)))
print("dawn:"+ str(count.get("dawn", 0)))
print("employment:"+ str(count.get("employment", 0)))
```

```python
#Question 10
with open("walden.txt", 'r', encoding='utf8') as input_file, open("cleaned_walden.txt", 'w', encoding='utf8') as output_file:
    contents = input_file.readlines()    
    for ind, val in enumerate(contents):
        contents[ind]=val.replace("\n", " ").replace(". ","\n").replace("! ","\n").replace("? ","\n").replace(",","").replace(":","").replace(";","").lower().strip("")
        output_file.write(contents[ind]) 
```
